```python
# importing required packages
import pandas as pd
```

```python
# importing csv file as dataframe using pandas
df = pd.read_csv("Corona.csv")
```

```python
df
```

```python
# displaying first 5 rows
df.head()
```

```python
# dataset information
df.info()
```

```python
# checking for missing values
df.isnull()
```

```python
# inserting the active cases column in df

df['Active']= df['Confirmed']-(df['Deaths'] + df['Recovered'])
df.head(10)
```

```python
# latest number of confirmed, deaths, recovered and active cases country wise

df.groupby(['Country/Region'])['Confirmed','Deaths','Recovered','Active'].sum()
```

```python
# latest number of confirmed, deaths and recovered cases Country/Region - Province/State wise

df.groupby(["Country/Region","Province/State"]) ["Confirmed","Recovered","Deaths"].sum()
```

```python
# Chinese province wise cases 

df.iloc[:,[1,2,4,6,7]][df['Country/Region'] == "China"]
```

```python
# top 10 countries of coronavirus data

country_cases = df.groupby('Country/Region')[['Confirmed','Recovered','Deaths']].sum()
country_cases.sort_values(by=['Confirmed'],ascending=False).head(10) 
```

```python
# countries where all cases of corona recovered

country_rec = df.groupby('Country/Region') [['Confirmed','Recovered']].sum() 
country_rec = country_rec[country_rec['Confirmed'] == country_rec['Recovered']]
country_rec = country_rec[country_rec["Confirmed"]>0]
country_rec
```
